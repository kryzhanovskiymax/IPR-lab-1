```python
import cv2
import matplotlib.pyplot as plt
from skimage.measure import label
import numpy as np
image_path = 'images/1.jpg'
image = cv2.imread(image_path)
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
def find_connected_components(image_path):
    # Загрузка изображения с использованием OpenCV
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Применение пороговой обработки, если необходимо
    _, thresholded_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    
    # Поиск связных компонентов
    labeled_image = label(thresholded_image)
    
    return labeled_image

# Пример использования функции
image_path = 'images/1.jpg'
labeled_image = find_connected_components(image_path)
# # Вывод изображения с найденными компонентами
plt.imshow(labeled_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
image_path = 'images/1.jpg'
image = cv2.imread(image_path)

image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, np.ones((4,4), np.uint8))
image = cv2.morphologyEx(image, cv2.MORPH_OPEN, np.ones((3, 3), np.uint8))


# Разделение каналов изображения
b, g, r = cv2.split(image)

# Создание маски, где пиксели с компонентой blue меньше 200 будут иметь значение 0, а остальные - 255
_, mask = cv2.threshold(b, 195, 255, cv2.THRESH_BINARY)

# Применение маски к изображению
result = cv2.bitwise_and(image, image, mask=mask)
# mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, np.ones((4,4), np.uint8))
# mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((3, 3), np.uint8))
countours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
count = 0
for contour in countours:
    if cv2.contourArea(contour) > 20:
        count += 1
        cv2.drawContours(result, [contour], -1, (0, 255, 0), 2)
print(f'Количество найденных объектов: {count}')

plt.imshow(result)
plt.axis("off")
plt.show()
```
